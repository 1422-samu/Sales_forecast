```python
import pandas as pd
```

```python
import pyodbc

# Define the connection parameters
server = r'LAPTOP-JLDIRAO0\SQLEXPRESS'
database = r'AdventureWorksDW2022'

# Establish the connection with Windows authentication
conn = pyodbc.connect('DRIVER={SQL Server};SERVER='+server+';DATABASE='+database+';Trusted_Connection=yes;')

# Create a cursor object to interact with the database
cursor = conn.cursor()
```

```python
try:
    query = """
    SELECT 
        fis.OrderDate,
        dc.CustomerKey,
        dc.BirthDate AS Age,
        dc.YearlyIncome AS Income,
        dc.Gender,
        dp.ProductKey,
        dp.ListPrice AS UnitPrice,
        dg.GeographyKey,
        dg.City,
        dg.CountryRegionCode,
        dprm.DiscountPct,
        fis.SalesAmount,
        fis.OrderQuantity
    FROM 
        FactInternetSales fis
    JOIN 
        DimCustomer dc ON fis.CustomerKey = dc.CustomerKey
    JOIN 
        DimProduct dp ON fis.ProductKey = dp.ProductKey
    JOIN 
        DimGeography dg ON fis.SalesTerritoryKey = dg.SalesTerritoryKey
    JOIN 
        DimPromotion dprm ON fis.PromotionKey = dprm.PromotionKey
    ORDER BY 
        fis.OrderQuantity DESC
    """

    # Execute the query
    cursor.execute(query)

    # Fetch results
    rows = cursor.fetchall()

    # Close cursor and connection
    cursor.close()
    conn.close()

except pyodbc.Error as e:
    print(f"Error connecting to the database: {e}")
```

```python
# Creating a dictionary of the Data

column_names = [
    'OrderDate',
    'CustomerKey',
    'Age',
    'Income',
    'Gender',
    'ProductKey',
    'UnitPrice',
    'GeographyKey',
    'City',
    'CountryRegionCode',
    'DiscountPct',
    'SalesAmount',
    'OrderQuantity'
]


dictionary = {i:[] for i in column_names}
dictionary

def conversion(X):
    for tup in X :
        i=0
        for key in dictionary.keys():
            dictionary[key].append(tup[i])
            i = i+1
    return dictionary

Data = conversion(rows)
```

```python
df = pd.DataFrame(Data)
```
